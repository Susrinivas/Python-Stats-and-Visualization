**6.Cleanse the data to make the year, population, cost, cost per person numerical type**

```python
def cleanse_year(a_string):
    return int(a_string.replace("*",""))
```

```python
df["Census Year"] = df["Census Year"].apply(cleanse_year)

df.tail()
```

```python
def cleanse_pop(a_string):
    return int(a_string.replace(",",""))
```

```python
df["Total Population"] = df["Total Population"].apply(cleanse_pop)

df.head()
```

```python
def cleanse_cost(a_string):
    cost1=a_string.replace("$","")
    cost2=cost1.replace(",","")
    a_string=int(float(cost2.replace(" Billion",""))*1000000000)
    return int(a_string)
```

```python
df["Census Cost"] = df["Census Cost"].apply(cleanse_cost)

df.head()
```

```python
def cleanse_avg_cost(avg_cost):
    avg_cost1=avg_cost.replace("$","")
    avg_cost2=avg_cost1.replace("\n","")
    if "cents" in avg_cost2:
         avg_cost3=float(avg_cost2.replace(" cents",""))/100
    else:
         avg_cost3=float(avg_cost2)
    r = round(avg_cost3, 5)
    return (r)
```

```python
df["Average Cost Per Person"] = df["Average Cost Per Person"].apply(cleanse_avg_cost)

df.head()
```

**8.Use Plotly to plot the average cost per person over time.**

```python
import plotly.express as px
```

```python
fig = px.line(
    df,
    x="Census Year",
    y="Average Cost Per Person", 
    title="the average cost per person over time.",  
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.show()
```

**9.Use Plotly to plot the scatter plot of Total Population vs Total Cost.**

```python
fig = px.scatter(
    df,
    x="Total Population",
    y="Census Cost", 
    title="the scatter plot of Total Population vs Total Cost",  
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```
