Exploratory data analysis on science and technology

In this project I am doing exploratory data ananlysis on science and technology using data from several countries mainly BRICS nations

```python
import pandas as pd
```

```python

df = pd.read_csv('https://raw.githubusercontent.com/Susrinivas/Python-Stats-and-Visualization/main/data690_world_dev/wdx_wide%20(4).csv')
df.shape #checking shape of dataframe
```

```python
df.head()
```

```python
df.tail()
```

```python
#renaming column names to readable format
df = df.rename(columns = {"GB.XPD.RSDV.GD.ZS":"Research and development expenditure (% of GDP)","SP.POP.TECH.RD.P6":"Technicians in R&D (per million people)","TX.VAL.TECH.CD":"High-technology exports (current US$)"})
df.head() 
```

```python
df.sample(5)
```

```python
df.describe() # to know certain statistical measures of data
```

```python
import plotly.express as px  # importing plotly
```

```python
df1=df[df['Country Name'].isin(['Brazil','China','India','Russian Federation','South Africa'])]
df1
```

```python
df2=df1[df['Year']==2020]
df2
```

```python
#Here is the bar chart of the data
fig = px.bar(df2, x = 'Country Name', y = 'High-technology exports (current US$)')
fig.show()
```

China exports are over a whooping 700 billion dollars in year 2020, even no country other than china is able to export to this quantity, and next comes india with around 50 billion dollars, but india has almost equal population as china, but how china is able to achieve success in this field, if you look back at 1990, india and china have equal resources, china communist party started implementing new economic polices and reforms which leads to present china where india lacks at bringing policies at right time So atleast governments should start enforcing new polices by taking china as an example

```python
fig = px.pie(df2, values='High-technology exports (current US$)', names='Country Name', title='pie chart which shows technology goods exporting countries')
fig.show()
```

The above pie chart have same data as bar chart, as you seen 95% of technology exports come from china in brics nations, ofcourse in military equipment Russia is the leading exporter

```python
fig=px.line(df1,x='Year',y='High-technology exports (current US$)',color='Country Name')
fig.show()
```

if you look at the above graph which depicts the historic trend shows that from 2007 china is like a f1 race car which has steep rise apart from some sudden breaks (ups and downs), and india is like a car moving with uniform speed, india and other nations should really concentrate on their policies

```python
df3=df[df['Year']==2018]
df3
```

```python
fig = px.scatter(
    df3,
    x="Technicians in R&D (per million people)",
    y="High-technology exports (current US$)", 
    color='Country Name',
    title="the scatter plot of Technicians in R&D vs exports ($ US)",  
)


fig.show()
```
