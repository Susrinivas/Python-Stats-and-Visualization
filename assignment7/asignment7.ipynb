```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv")

df.shape
```

```python
df.sample(5)
```

```python
df["indicator"].unique()
```

```python
df1 = df.query('indicator == "NY.GDP.PCAP.PP.CD" ')    #considering GDP per capita indicator

df1
```

```python
df["Country Name"].unique()
```

```python
my_countries = ["France", "Ukraine", "South Africa","Saudi Arabia","Pakistan"]

df2 = df1[df1["Country Name"].isin(my_countries)]

df2.shape
```

```python
df3 = df2.query(' Year == 2021')

df3
```

```python
fig = px.bar(
    df3, 
    x="Country Name", 
    y="value", 
    title="2021 GDP percapita of France,Ukraine,South Africa,Saudi Arabia,Pakistan ",
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```

```python
fig = px.pie(
    df3, 
    names="Country Name", 
    values="value", 
    title="2021 GDP percapita of France,Ukraine,South Africa,Saudi Arabia,Pakistan", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

```python
fig = px.box(df3, x="Country Name", y="value")
fig.show()
```

```python
df5 = df2.groupby(['Region', 'Year']).agg({'value': ['mean']})
df5.columns = ['value_mean']
df5 = df5.reset_index()
print(df5)
```

https://jamesrledoux.com/code/group-by-aggregate-pandas

```python
fig = px.line(
    df5,
    x="Year",
    y="value_mean", 
    title="2021 GDP percapita of France,Ukraine,South Africa,Saudi Arabia,Pakistan", 
    color="Region", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.show()
```

```python
df6 = df.query(' Year == 2020')
df6
```

```python
my_indicators = ["SP.DYN.LE00.IN", "NY.GDP.PCAP.PP.CD"]

df7 = df6[df6["indicator"].isin(my_indicators)]

df7.shape
```

```python
df8 = pd.pivot_table(df7, values = 'value', index=['Country Code', 'Country Name','Year'], columns = 'indicator').reset_index()

df8.head()
```

```python
df8.dropna(inplace=True)

df8.head()
```

```python
fig = px.scatter(
    df8,
    x="NY.GDP.PCAP.PP.CD",
    y="SP.DYN.LE00.IN", 
    title="Wealth vs Health", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```
