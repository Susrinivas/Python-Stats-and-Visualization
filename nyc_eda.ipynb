```python
df2['year']=df.pickup_datetime.dt.year
df2['month']=df.pickup_datetime.dt.month
df2['day']=df.pickup_datetime.dt.day
df2['weekday']=df.pickup_datetime.dt.weekday
df2['hour']=df.pickup_datetime.dt.hour
```

```python
import numpy as np
from math import radians, cos, sin, asin, sqrt

def haversine_np(lon1, lat1, lon2, lat2):
   # convert decimal degrees to radians 
   lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

      # haversine formula 
   dlon = lon2 - lon1 
   dlat = lat2 - lat1 
   a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
   c = 2 * np.arcsin(np.sqrt(a)) 
   r = 6367 # Radius of earth in kilometers.
   return c * r
```

```python
df2['distance']=haversine_np(df2['pickup_longitude'],df2['pickup_latitude'],df2['dropoff_longitude'],df2['dropoff_latitude'])
```
